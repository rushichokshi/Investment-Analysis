```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
```

```python
bitcoin = pd.read_csv('../data/bitcoin_daily.csv')
```

```python
bitcoin = bitcoin.drop(columns=['Open','High','Low','Adj Close','Volume'],axis=1)
```

```python
########################################################################################################
```

```python
bitcoin.plot(x="Date",y="Close",kind="line")
plt.xticks(rotation=45)
plt.show()
```

Determine standard deviation for the last 3 years, 5 years, and 10 years.

3 years

```python
print(statistics.stdev(bitcoin[bitcoin['Date']>='2021-02-23']['Close']))
print(bitcoin[bitcoin['Date']>='2021-02-23']['Close'].std())
```

5 years

```python
print(statistics.stdev(bitcoin[bitcoin['Date']>='2019-02-20']['Close']))
print(bitcoin[bitcoin['Date']>='2019-02-20']['Close'].std())
```

10 years

```python
print(statistics.stdev(bitcoin[bitcoin['Date']>='2014-02-20']['Close']))
print(bitcoin[bitcoin['Date']>='2014-02-20']['Close'].std())
```

```python
sns.lineplot(x=bitcoin[bitcoin['Date']>='2020-02-01'][bitcoin['Date']<='2020-04-30']['Date'],y=bitcoin[bitcoin['Date']>='2020-02-01'][bitcoin['Date']<='2020-04-30']['Close'])
plt.grid(True)
plt.xticks(rotation=90)
plt.xticks(ticks=['2020-02-01','2020-04-30','2020-02-12','2020-03-27','2020-03-11'],fontsize=8)

plt.scatter('2020-02-12',10326.054688,marker='o',color='red')
plt.text('2020-02-12',10326.054688 + 100,'2/12 $10,326.05',fontsize=8)

plt.scatter('2020-03-11',7911.430176,marker='o',color='red')
plt.text('2020-03-11',7911.430176 + 200,'3/11 $7,911.43',fontsize=8)

plt.scatter('2020-03-12',4970.788086,marker='o',color='red')
plt.text('2020-03-12',4970.788086 - 225,'3/12 $4,970.79',fontsize=8)

plt.scatter('2020-03-27',6469.79834,marker='o',color='red')
plt.text('2020-03-12',6469.79834 + 350,'3/27 $6,469.80',fontsize=8)

plt.yticks(fontsize=8)
plt.title('Bitcoin Feb-Apr 2020')

plt.show()
```
