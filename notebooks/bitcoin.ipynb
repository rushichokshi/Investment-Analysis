```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
```

```python
bitcoin = pd.read_csv('../data/bitcoin_daily.csv')
```

```python
bitcoin = bitcoin.drop(columns=['Open','High','Low','Adj Close','Volume'],axis=1)
```

```python
########################################################################################################
```

```python
bitcoin.plot(x="Date",y="Close",kind="line")
plt.xticks(rotation=45)
plt.ylabel('Price')
plt.legend().remove()
plt.title('Bitcoin 2014-present')
plt.show()
```

```python
########################################################################################################
```

Determine standard deviation for the last 3 years, 5 years, and 10 years.

3 years

```python
print(statistics.stdev(bitcoin[bitcoin['Date']>='2021-03-01']['Close']))
print(bitcoin[bitcoin['Date']>='2021-03-01']['Close'].std())
```

5 years

```python
print(statistics.stdev(bitcoin[bitcoin['Date']>='2019-03-01']['Close']))
print(bitcoin[bitcoin['Date']>='2019-03-01']['Close'].std())
```

10 years

```python
print(statistics.stdev(bitcoin[bitcoin['Date']>='2014-03-01']['Close']))
print(bitcoin[bitcoin['Date']>='2014-03-01']['Close'].std())
```
