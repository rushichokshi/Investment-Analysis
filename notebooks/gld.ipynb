```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
```

```python
gld = pd.read_csv('../data/gld_daily.csv')
```

```python
gld = gld.drop(columns=['Open','High','Low','Close','Volume'],axis=1)
```

```python
gld.plot(x="Date",y="Adj Close",kind="line")
plt.xticks(rotation=45)
plt.title('GLD ETF 2004-present')
plt.legend().remove()
plt.ylabel('Price')
plt.show()
```
