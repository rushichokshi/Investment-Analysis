How did GLD ETF react to COVID-19?

```python
sns.lineplot(x=gld[gld['Date']>='2020-02-01'][gld['Date']<='2020-04-01']['Date'],y=gld[gld['Date']>='2020-02-01'][gld['Date']<='2020-04-01']['Adj Close'])
xticks_to_keep = ['2020-03-01', '2020-03-11', '2020-03-12', '2020-03-31']
plt.xticks(rotation=45,ticks=xticks_to_keep)
plt.grid(True)
plt.show()
```

```python
sns.lineplot(x=gld[gld['Date']>='2020-03-01'][gld['Date']<='2020-03-31']['Date'],y=gld[gld['Date']>='2020-03-01'][gld['Date']<='2020-03-31']['Adj Close'])
plt.grid(True)
plt.xticks(rotation=90)
point_to_label=('2020-03-11',153.929993)
label_text='Pandemic announcement, $153.93'
plt.annotate(label_text,xy=point_to_label,xytext=('2020-03-15',15),textcoords='offset points',arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=0.5',color='red'),fontsize=8)
plt.show()
```
